```python
import json
import os
import pandas as pd
from pathlib import Path
from tqdm import tqdm
```

```python
# paths to sampled stories
data_dir = Path('../data/sampled_60')  # adjust if your data is elsewhere
stories_20_path = data_dir / 'sampled_20_stories.json'
stories_40_path = data_dir / 'sampled_40_stories.json'

# output directories for HTML files
output_dir_20 = Path('./data/input_amt_samples_20')
output_dir_40 = Path('./data/input_amt_samples_40')

# task template
template_path = Path('./data/properties/task-template.html')

# create output directories
output_dir_20.mkdir(parents=True, exist_ok=True)
output_dir_40.mkdir(parents=True, exist_ok=True)

print(f"Stories (20): {stories_20_path}")
print(f"Stories (40): {stories_40_path}")
print(f"Template: {template_path}")
print(f"Output (20): {output_dir_20}")
print(f"Output (40): {output_dir_40}")
```

```python
def is_valid_value(value):
    """check if value is valid (not None, NaN, empty, or '{}')."""
    if value is None:
        return False
    if pd.isna(value):
        return False
    value_str = str(value).strip()
    if value_str == '' or value_str.lower() == 'nan' or value_str == '{}':
        return False
    return True


def process_story_images(row):
    """extract valid story images from a row."""
    images = []
    for num in range(10):
        col_name = f'link{num}'
        if col_name in row and is_valid_value(row[col_name]):
            images.append({'ImageSrc': row[col_name], 'ImgId': num})
    return images


def process_character_images(row):
    """extract valid character images and names from a row."""
    characters = []
    for num in range(5):
        url_col = f'char{num}_url'
        name_col = f'char{num}'
        if (url_col in row and name_col in row and 
            is_valid_value(row[url_col]) and is_valid_value(row[name_col])):
            characters.append({
                'CharImageSrc': row[url_col],
                'CharNameSrc': row[name_col]
            })
    return characters


def generate_html_files(stories_path, output_dir, template_path, desc):
    """generate HTML files for a set of stories."""
    # load stories
    with open(stories_path, 'r', encoding='utf-8') as f:
        stories = [json.loads(line) for line in f if line.strip()]
    
    # load template
    with open(template_path, 'r', encoding='utf-8') as f:
        html_template = f.read()
    
    # generate HTML for each story
    for story in tqdm(stories, desc=desc):
        story_id = story['story_id']
        
        # process images and characters
        story_images = process_story_images(story)
        char_images = process_character_images(story)
        
        print(f"Story {story_id}: {len(story_images)} images, {len(char_images)} characters")
        
        # fill template
        html_content = html_template.replace('IMAGESTOINSERT', str(story_images))
        html_content = html_content.replace('CHARSTOINSERT', str(char_images))
        html_content = html_content.replace('STORYID', str(story_id))
        
        # save HTML file
        output_file = output_dir / f'{story_id}.html'
        with open(output_file, 'w', encoding='utf-8') as f:
            f.write(html_content)
    
    print(f"\nGenerated {len(stories)} HTML files in {output_dir}")
```

```python
generate_html_files(
    stories_path=stories_20_path,
    output_dir=output_dir_20,
    template_path=template_path,
    desc="Generating HTMLs for 20 stories"
)
```

```python
generate_html_files(
    stories_path=stories_40_path,
    output_dir=output_dir_40,
    template_path=template_path,
    desc="Generating HTMLs for 40 stories"
)
```
